5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 1000

# Generate synthetic data with correlated features
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix indicating correlation between features
data = np.random.multivariate_normal(mean, cov, n_samples)

# Convert data to DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Visualize correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade
categories

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Student Grades data
np.random.seed(15)
grade_data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
grades_df = pd.DataFrame(grade_data)

# Sales data
np.random.seed(20)
sales_data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
sales_df = pd.DataFrame(sales_data)

# Create a violin plot for student grades
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', data=grades_df, order=['A', 'B', 'C', 'D', 'F'])
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()

# Create a heatmap for sales data
sales_pivot = sales_df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc=np.mean)
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, cmap='YlGnBu')
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```

5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 8))
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.5)

# Add labels and title
plt.title('Bubble Chart of GDP vs Population')
plt.xlabel('GDP')
plt.ylabel('Population')

# Add country labels
for i, txt in enumerate(df['Country']):
    plt.annotate(txt, (df['GDP'][i], df['Population'][i]), fontsize=8, ha='right')

# Set x-axis and y-axis limits for better visualization
plt.xlim(0, max(df['GDP']) * 1.1)
plt.ylim(0, max(df['Population']) * 1.1)

plt.grid(True)
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
output_file("sine_wave.html")
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.




```python
from bokeh.plotting import figure, output_file, show
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
n_points = 100
data = {
    'x': np.random.randint(0, 100, n_points),
    'y': np.random.randint(0, 100, n_points),
    'sizes': np.random.randint(5, 15, n_points),
    'colors': np.random.choice(['red', 'blue', 'green'], n_points)
}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a Bokeh figure
output_file("scatter_plot.html")
p = figure(title="Scatter Plot")

# Add circles for each data point
p.circle(x='x', y='y', size='sizes', color=factor_cmap('colors', palette=['red', 'blue', 'green'], factors=['red', 'blue', 'green']), source=source)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
import pandas as pd

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = {'fruits': fruits, 'counts': counts}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a Bokeh figure
output_file("bar_chart.html")
p = figure(x_range=fruits, title="Counts of Different Fruits", x_axis_label='Fruits', y_axis_label='Counts')

# Add bars
p.vbar(x='fruits', top='counts', width=0.5, color='skyblue', legend_label='Fruit Counts', source=source)

# Set the y-axis to start at 0
p.y_range.start = 0

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
output_file("histogram.html")
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import LinearColorMapper
import numpy as np

# Given dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh figure
output_file("heatmap.html")
p = figure(title="Heatmap")

# Convert the data to a 1D array
data_flat = data_heatmap.flatten()

# Create a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())

# Add the heatmap
p.rect(x=xx.flatten(), y=yy.flatten(), width=1/10, height=1/10, fill_color={'field': 'values', 'transform': color_mapper},
       line_color=None, source={'x': xx, 'y': yy, 'values': data_heatmap})

# Show the plot
show(p)
```
